```python
#import liberary
import numpy as np
import pandas as pd
```

```python
#import file csv
data = pd.read_csv('Automobile.csv')
data.dropna(inplace=True)
#choose columns
data=data[['company','wheel-base','length','horsepower','average-mileage','price']]
data.head()
```

```python
#crate class to convert dataframe to numpy array
class Con:
    def __init__(self,value):
        self.value = value
    
    def convert(self):
        self.df =pd.DataFrame(self.value)
        self.df =self.df.to_numpy()
        return self.df

# create class to calculate min-max and z-score normalization
class Nor:
    def __init__(self,num):
        self.num = num

    def minmax(self): #min-max normalization
        min_max=[]
        min = np.min(self.num)
        max = np.max(self.num)
        
        for i in self.num:
            self.nor = (i-min)/(max-min)
            min_max.append(self.nor)
            
        return min_max
    
    def zscore(self): #z-score normalization
        mean =np.mean(self.num)
        std =np.std(self.num)
        outline =[]
        for i2 in self.num:
            self.ans=(i2-mean)/std
            outline.append(self.ans)
        return outline

# create class to convert numpy array to dataframe
class Get:
    
    def __init__(self,normal,score):
        self.normal = normal
        self.score = score
    
    def create(self):
        dataframe = {'min-max':self.normal,'zscore':self.score}
        df_new = pd.DataFrame(dataframe)
        print(df_new)
        
```

```python
# create function to get result 
def fw(x):
    result = Con(x)
    normal =Nor(result.convert())
    frame=Get(normal.minmax(),normal.zscore())
    com = frame.create()
    combine = pd.concat([data['company'],x,com], axis=1, join='inner')
    print(combine)
    
wheel=fw(data['wheel-base'])
```
